# **File Handling**

<p align="justify">File is a named location on disk to store related information. It is used to permanently store data in a non-volatile memory (e.g. hard disk).

<p align="justify">Since, random access memory (RAM) is volatile which loses its data when computer is turned off, we use files for future use of the data.

<p align="justify">Hence, in Python, a file operation takes place in the following order.
1.	Open a file
2.	Read or write (perform operation)
3.	Close the file


# **a).How to open a file**

<p align="justify">Python has a built-in function open() to open a file. This function returns a file object, also called a handle, as it is used to read or modify the file accordingly.

# **File handling methods:**

# **a).read()**

<p align="justify">The read() method returns the specified number of bytes from the file. Default is -1 which means the whole file.

# **b).readline()**

<p align="justify">The readline() method returns one line from the file.You can also specified how many bytes from the line to return, by using the size parameter.

# **c).readlines()**

<p align="justify">The readlines() method returns a list containing each line in the file as a list item.

<p align="justify">If parameter is passed i will read and returns a list of the line/lines corrosponding to that bytes. If the total number of bytes returned exceeds the specified number, no more lines are returned.

# **d).seek()**

<p align="justify">The seek() method sets the current file position in a file stream.

# **e).tell()**

<p align="justify">The tell() method returns the current file position in a file stream.

# **f).truncate()**


<p align="justify">The truncate() method resizes the file to the given number of bytes. If the size is not specified, the current position will be used.

# **g).write()**

<p align="justify">The write() method writes a specified text to the file.

# **h).writelines()**

<p align="justify">The writelines() method writes the items of a list to the file.

# **i).flush()**

<p align="justify">The flush() method cleans out the internal buffer.

<p align="justify">Python automatically flushes the files when closing them. But you may want to flush the data before closing any file.
